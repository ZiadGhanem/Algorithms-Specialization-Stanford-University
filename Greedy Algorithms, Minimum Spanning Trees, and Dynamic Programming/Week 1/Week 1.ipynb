```
# Files path
JOBS_FILE_PATH = "/content/Algorithms/Course 3/Week 1/jobs.txt"
EDGES_FILE_PATH = "/content/Algorithms/Course 3/Week 1/edges.txt"
```

```
jobs_list = list()
with open(JOBS_FILE_PATH) as jobs_file:
  jobs = jobs_file.readlines()
  for index, job in enumerate(jobs[1:]):
    jobs_list.append([int(item) for item in job.split()])
    jobs_list[index].append(jobs_list[index][0] - jobs_list[index][1])
    jobs_list[index].append(jobs_list[index][0] / jobs_list[index][1])
```

```
print(jobs_list)
print(len(jobs_list))
```

```
jobs_list = sorted(jobs_list, key = lambda x: (x[2], x[0]), reverse=True)
```

```
print(jobs_list)
```

```
difference_completion_time = 0
difference_total_weight = 0
for job in jobs_list:
  difference_completion_time += job[1]
  difference_total_weight += job[0] * difference_completion_time
print("The sum of weighted completion times= " + str(difference_total_weight))
```

```
jobs_list = sorted(jobs_list, key = lambda x: x[3], reverse=True)
print(jobs_list)
```

```
ratio_completion_time = 0
ratio_total_weight = 0
for job in jobs_list:
  ratio_completion_time += job[1]
  ratio_total_weight += job[0] * ratio_completion_time
print("The sum of weighted completion times= " + str(ratio_total_weight))
```

```
edges_list = list()
with open(EDGES_FILE_PATH) as edges_file: 
  edges = edges_file.readlines()
  number_nodes = int((edges[0].split())[0])
  for edge in edges[1:]:
    edges_list.append([int(item) for item in edge.split()])
```

```
print(number_nodes)
print(edges_list)
```

```
edges_list = sorted(edges_list, key = lambda x: x[0])
print(edges_list)
```

```
adjacency_dict = dict()
for edge in edges_list:
  if edge[0] not in adjacency_dict.keys():
    adjacency_dict[edge[0]] = [edge[1:]]
  else:
    adjacency_dict[edge[0]].append(edge[1:])
for edge in edges_list:
  if edge[1] not in adjacency_dict.keys():
    adjacency_dict[edge[1]] = [[edge[0], edge[2]]]
  else:
    adjacency_dict[edge[1]].append([edge[0], edge[2]])
```

```
print(adjacency_dict)
```

```
visited_nodes = list()
visited_nodes.append(1)
minimum_spanning_tree = dict()
total_cost = 0
while len(visited_nodes) != number_nodes:
  min_edge_cost = float('inf')
  for node in adjacency_dict.keys():
    if node in visited_nodes:
      for neighbour_node, cost in adjacency_dict[node]:
        if neighbour_node not in visited_nodes:
          if cost < min_edge_cost:
            min_edge_cost = cost
            min_edge = neighbour_node
  visited_nodes.append(min_edge)
  total_cost += min_edge_cost
```

```
print("Overall cost of a minimum spanning tree= " + str(total_cost))
```
