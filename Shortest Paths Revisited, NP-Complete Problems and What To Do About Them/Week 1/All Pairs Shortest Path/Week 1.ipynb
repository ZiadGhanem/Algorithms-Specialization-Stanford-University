```
import numpy as np
from tqdm import tqdm
```

```
def read_file(graph_file_path):
  with open(graph_file_path) as graph_file:
    edges = graph_file.readlines()
    # Read number of vertices and edges
    number_vertices = int((edges[0].split())[0])
    number_edges = int((edges[0].split())[1])

    A = np.zeros((number_vertices, number_vertices, number_vertices + 1))
    inf = 9999
    for i in range(number_vertices):
      for j in range(number_vertices):
        if i == j:
          A[i, j, 0] = 0
        else:
          A[i, j, 0] = inf

    for index, edge in enumerate(edges[1:]):
      items = edge.split()
      A[int(items[0]) - 1, int(items[1]) - 1, 0] = int(items[2])

    return A
```

```
def floyd_warshall(A):
  number_vertices = A.shape[0]
  for k in tqdm(range(1, number_vertices + 1)):
    for i in range(number_vertices):
      for j in range(number_vertices):
        A[i, j, k] = min(A[i, j, k-1], A[i, k - 1, k - 1] + A[k - 1, j, k-1])

  for i in range(number_vertices):
    if A[i, i, number_vertices] < 0:
      return False

  return np.min(A[:, :, number_vertices])
```

```
for graph_index, graph_path in enumerate(GRAPH_PATHS):
  A = read_file(graph_path)
  result = floyd_warshall(A)
  if result != False:
    print("Shortest path in graph " + str(graph_index) + "= " + str(result))
  else:
    print("Graph " + str(graph_index) + " contains negative cycles")
```
